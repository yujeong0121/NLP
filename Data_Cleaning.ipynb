```python
# 정규표현식으로 패턴 제거
def preprocess_text(text):
    
    text = re.sub('제.조', '', text) # 제1조, 제2조, ... 
    text = re.sub('제..조', '', text) # 제10조, 제11조, ... 
    text = re.sub('제...조', '', text) # 제100조, 제101조, ...
    text = re.sub('제.항','',text) # 제1항, 제2항, ...
    text = re.sub('제..항','',text) # 제10항, 제11항, ...
    text = re.sub('제.호','',text) # 제1호, 제2호, 제3호, ...
    text = re.sub('제..호','',text) # 제10호, 제11호, ...
    text = re.sub('[①-⑳]','',text) # ①, ②, ③, ... 
    text = re.sub('[\t\n]','',text) # 공백 제거
    text = re.sub('[-=+,#/\?:^$@*\"※~&%ㆍ!』「」\\‘|\(\)\[\]`\'…》]','',text) # 특수문자 제거
    text = re.sub('[/一-龥/]', '', text) # 한자 제거
    text = re.sub('<.*?>','', text) # <개정 2022.4.23>, < >와 < > 사이에 있는 문자 제거
    text = re.sub('[0-9][.]','', text) # 숫자. 제거
    text = re.sub('[.]','',text) # .(마침표) 제거
    
    return text
```

```python
law_data['조문내용'] = law_data['조문내용'].apply(preprocess_text)
```
